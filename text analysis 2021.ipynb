```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import PyPDF2
```

```python
file_path= 'Hutbeler/2021 Yılı Türkçe Hutbeleri.pdf'
```

# Preprocessing I

```python
# creating a pdf file object
pdfFileObj = open(file_path, 'rb')
# creating a pdf reader object
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
# printing number of pages in pdf file
print(pdfReader.numPages)
# creating a page object
pageObj = pdfReader.getPage(0) #first page = 0
# extracting text from page
print(pageObj.extractText())
```

```python
#create a list that seperates pages.

hutbe2021=[]

for page in range(pdfReader.getNumPages()):
    current_page = pdfReader.getPage(page)
    pdf=current_page.extractText()
    hutbe2021.append(pdf)
```

```python
hutbe2021[0] #first hutbe in the list
```

```python
#create a dataframe from the list
hutbe_2021_df=pd.DataFrame(hutbe2021, columns =['content'])

hutbe_2021_df.head()
```

```python
#there was a date missing at row 28. I added manually

hutbe_2021_df.content.iloc[28]='Tarih: 16.07.2021' + hutbe_2021_df.content.iloc[28]
```

```python
#replace all \n
hutbe_2021_df['content']=hutbe_2021_df.content.str.replace('\n', '')

hutbe_2021_df.head()
```

```python
#use regex to extract date because we want to create a date column

pattern= '\d{2}.\d{2}.\d{4}'
text=str(hutbe_2021_df.content)

date=re.findall(pattern, text)

len(date) #make sure it's len match. 
```

```python
#here is the date we extracted from all hutbe
date[:5]
```

```python
#let's drop "Tarih" not part of the text. 

hutbe_2021_df['content']=hutbe_2021_df.content.str.lstrip('Tarih:')

hutbe_2021_df.head()
```

```python
#some words are cut in the middle.

pattern=(r'\D{8} -\D{8}')

text= str(hutbe_2021_df.content)

re.findall(pattern, text)
```

```python
#i did my best to fix some words. 

hutbe_2021_df['content']=hutbe_2021_df.content.str.replace('ayet-i', 'ayeti')
hutbe_2021_df['content']=hutbe_2021_df.content.str.replace('Mevlid-i N', 'mevlidin')
hutbe_2021_df['content']=hutbe_2021_df.content.str.replace('millet  -i', 'milleti')

hutbe_2021_df.head()
```

```python
#After extracting date from the text, now we can continue doing preprocessing
# such as drop all the number etc...

import re
import string

def clean_text_round1(text):
    '''Make text lowercase, 
    remove text in square brackets,
    remove punctuation
    remove words containing numbers.'''

    text = text.lower()
    text = re.sub('\[.*?\-""''..._]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
# Let's take a look at the updated text
hutbe_21_round1=pd.DataFrame(hutbe_2021_df.content.apply(round1))

hutbe_21_round1.head()
```

```python
#lets replace standalone words with nothing. 
 
pattern= ' [a-zA-Z] '

result=re.sub(pattern, '', text)

pd.Series(result)
```

# Preprocessing II

```python
clean_2021= hutbe_21_round1.copy()

clean_2021.head()
```

```python
#stopwords in turkish
import advertools as adv

sw= adv.stopwords['turkish']

#sw.append('bu')

sw=list(sw)
```

```python
clean_2021['content']=clean_2021['content'].apply(lambda x: ' '.join(x for x in x.split() if x not in sw))

clean_2021.head()
```

```python
### most words frequency
pd.Series(' '.join(clean_2021.content).split()).value_counts()
```

```python
more_sw=pd.Series(" ".join(clean_2021.content).split()).value_counts()[:2]
#or least words [-1:-6]
```

```python
#delete more stop words
clean_2021['content']=clean_2021['content'].apply(lambda x: ' '.join(x for x in x.split() if x not in more_sw))

clean_2021.head()
```

```python
pd.Series(" ".join(clean_2021.content).split()).value_counts()
```

### tokanization

```python
#second way tokenization
import nltk
nltk.download('punkt')

from nltk.tokenize import sent_tokenize

tokened=clean_2021.content.apply(lambda x: sent_tokenize(x))
```

```python
clean_2021['tokened']=tokened

clean_2021
```

## Lemmatization

```python
import zeyrek
analyzer = zeyrek.MorphAnalyzer()
```

```python

def lemmatization(token_txt):
    text= [analyzer.lemmatize(word) for word in token_txt]
    return text
```

```python
lemt=clean_2021['tokened'].apply(lambda x: lemmatization(x))
```

```python
clean_2021['lematized']=lemt #lematization
```

```python
clean_2021['lematized_v2']=lemt.str[0].explode().str[1].str[0].groupby(level=0).apply(",".join)

clean_2021.head()
```

```python
clean_2021.lematized_v2 =clean_2021.lematized_v2.str.lower() #lower case

clean_2021.head()
```
